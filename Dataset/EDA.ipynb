## Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
from plotnine import *
```

```python
Train=pd.read_csv('cs-training.csv').drop('Unnamed: 0',axis=1)
```

```python
Train
```

```python
# Exploratory Data Analysis
var=['RevolvingUtilizationOfUnsecuredLines',
     'age',
     'DebtRatio',
     'MonthlyIncome']
Train[var].describe()
```

```python
fig,ax=plt.subplots(2,2,figsize=(20,13))
ax[0,0].hist(Train[var[0]])
ax[0,0].title.set_text(var[0])
ax[0,1].hist(Train[var[1]])
ax[0,1].title.set_text(var[1])
ax[1,0].hist(Train[var[2]])
ax[1,0].title.set_text(var[2])
ax[1,1].hist(Train[var[3]])
ax[1,1].title.set_text(var[3])
#fig.suptitle('Distributions of the four parameters')
fig.savefig('Stats.png')
```
